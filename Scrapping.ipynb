```python
def get_path_to_file(p_filename):
    return folder + p_filename + ".xlsx"

def create_excel_file(p_file_name):
    if not os.path.isfile(p_file_name):
        workbook = xlsxwriter.Workbook(p_file_name)
        workbook.close()

def save_in_file(p_file_name, p_sheet_name, p_df):
    create_excel_file(get_path_to_file(p_file_name))

    writer = pd.ExcelWriter(get_path_to_file(p_file_name))
    p_df.to_excel(writer, p_sheet_name)
    writer.save()

def get_request_wait(p_link):
    try:
        req = requests.get(p_link)
        if req.status_code == 200:
            content = req.content

        return BeautifulSoup(content, 'html.parser'), True
    except Exception as e:
        print(e)
        randSleep = random.randint(500, 1500)/100
        print('Will rerty in %s seconds.' %(int(randSleep)))
        time.sleep(randSleep)
        return None, False

def get_request(p_link):
    success = False

    while success == False:
        req, success = get_request_wait(p_link)

    if success == True:
        return req

def get_header_from_link(p_link):
    soup = get_request(p_link)
    div = soup.find(id = "meta")
    h1 = div.find("h1")

    return h1.get_text().strip()

def get_subheader_from_link(p_link):
    soup = get_request(p_link)
    div = soup.find(id = "meta")
    h1 = div.find("h1")
    span = h1.find_all("span")

    return span[1].get_text().strip()
```

```python
def wcq_results(p_link):

    try:
        soup = get_request(p_link)
        tb = soup.find("table", id=re.compile('^results\d+_overall'))
        tbody = tb.find("tbody")

        wcq = get_header_from_link(p_link)

        results = []
        pos = 1

        for tr in tbody.find_all("tr"):
            result = WCQ_Results()

            for d in tr.find_all("td"):

                if hasattr(result, d['data-stat']):
                    if(d['data-stat'] == 'squad'):
                        a = d.find("a")
                        result.squad_detail_link = a["href"]
                        setattr(result, d['data-stat'], a.get_text().strip())
                    else:
                        setattr(result, d['data-stat'], d.get_text().strip()) 

                result.wcq = wcq
                result.rank = pos
            pos += 1
            results.append(result)

        return pd.DataFrame([r.__dict__ for r in results])

    except Exception as e:
        print("error:", e)
        exception_type, exception_object, exception_traceback = sys.exc_info()
        filename = exception_traceback.tb_frame.f_code.co_filename
        line_number = exception_traceback.tb_lineno

        print("Exception type: ", exception_type)
        print("File name: ", filename)
        print("Line number: ", line_number)

def wcq_squad_stats(p_df):
    try:
        stats = ["^stats_standard_\d+", "^stats_keeper_\d+","^stats_playing_time_\d+"]

        link = ""
        squads = []
        matches = []

        for index, row in p_df.iterrows():
            link = url_base + row["squad_detail_link"]
            soup = get_request(link)
            header = get_header_from_link(link)

            if header.find("WCQ") < 0 and header.find("2022") < 0: 
                div = soup.find("div", class_="filter")
                a = div.find_all("a")
                link = url_base + a[2]["href"]
                soup = get_request(link)
                header = get_header_from_link(link)
            
            for s in stats:
                tb = soup.find("table", id=re.compile(s))
                tbody = tb.find("tbody")

                for tr in tbody.find_all("tr"):
                    
                    squad = Squad_Stats()
                    squad.wcq = row["wcq"]
                    squad.squad = row["squad"] 
                    squad.squad_stats_link = link

                    for h in tr.find_all("th"):
                        if hasattr(squad, h['data-stat']):
                            setattr(squad, h['data-stat'], h.get_text().strip()) 
                    
                    if len(squads) > 0:
                        pos = 0
                        for i in squads:
                            if i.wcq == squad.wcq and i.squad == squad.squad and i.player == squad.player:
                                squad = i
                                squads.pop(pos)
                            pos += 1

                    for d in tr.find_all("td"):
                        if hasattr(squad, d['data-stat']):
                            setattr(squad, d['data-stat'], d.get_text().strip()) 

                    squads.append(squad)

            # get list of matches

            tb = soup.find("table", id="matchlogs_for")
            tbody = tb.find("tbody")

            for tr in tbody.find_all("tr"):
                flag_include = 1
                match = Matches_Scores()

                match.wcq = row["wcq"]

                for h in tr.find_all("th"):
                    if hasattr(match, h['data-stat']):
                        setattr(match, h['data-stat'], h.get_text().strip()) 
                    
                if len(matches) > 0:
                    pos = 0
                    for i in matches:
                        if i.wcq == match.wcq and i.date == match.date and (i.squad_home == row["squad"] or i.squad_away == row["squad"]):
                            match = i
                            matches.pop(pos)
                        pos += 1

                posfixo = ""
                for d in tr.find_all("td"):

                    if (d['data-stat'] == "round" and d.get_text().find("WCQ") < 0) or (d['data-stat'] == 'notes' and d.get_text() == "Match Suspended"):
                        flag_include = 0
                    
                    if(d['data-stat'] == 'match_report'):
                        a = d.find("a")
                        match.match_report_link = a["href"]

                        if a.get_text() != "Match Report":
                            flag_include = 0
                    
                    if d['data-stat'] == "venue" and d.get_text() == "Home":
                        posfixo = "_home"
                        match.squad_home = row["squad"]

                    if d['data-stat'] == "venue" and d.get_text() == "Away":
                        posfixo = "_away"
                        match.squad_away = row["squad"]

                    if d['data-stat'] == "opponent":
                        a = d.find("a")

                        if posfixo == "_home":
                            match.squad_away = a.get_text()
                        else:
                            match.squad_home = a.get_text()

                    match.squad_home_away = str(match.squad_home) + " - " + str(match.squad_away) 

                    if hasattr(match, d['data-stat'] + posfixo):
                        setattr(match, d['data-stat'] + posfixo, d.get_text().strip()) 
                        
                    if hasattr(match, d['data-stat'] ):
                        setattr(match, d['data-stat'] , d.get_text().strip()) 

                if flag_include == 1:
                    if match.goals_for_away == match.goals_for_home:
                        match.result_home = "D"
                        match.result_away = "D"
                        match.points_home = 1
                        match.points_away = 1
                    elif int(match.goals_for_away) > int(match.goals_for_home):
                        match.result_home = "L"
                        match.result_away = "W"
                        match.points_home = 0
                        match.points_away = 3
                    else:
                        match.result_home = "W"
                        match.result_away = "L"
                        match.points_home = 3
                        match.points_away = 0

                    matches.append(match)

        return pd.DataFrame([r.__dict__ for r in squads]), pd.DataFrame([r.__dict__ for r in matches])

    except Exception as e:
        print("error:", e)
        print("link:", link)
        exception_type, exception_object, exception_traceback = sys.exc_info()
        filename = exception_traceback.tb_frame.f_code.co_filename
        line_number = exception_traceback.tb_lineno

        print("Exception type: ", exception_type)
        print("File name: ", filename)
        print("Line number: ", line_number)

def wcq_match_stats(p_df):

    try:
        matches = []

        for index, row in p_df.iterrows():
            link = url_base + row["match_report_link"]

            soup = get_request(link)
            
            tb_p = soup.find_all("table", id=re.compile("^stats_\w+_summary"))
            tb_g = soup.find_all("table", id=re.compile("^keeper_stats_\w+"))

            if len(tb_p) == len(tb_g) and len(tb_p) == 2 > 0:
                for i in range(2):

                    tbody_p = tb_p[i].find("tbody")
                    tbody_g = tb_g[i].find("tbody")

                    for tr in tbody_p.find_all("tr"):

                        squad = Match_Stats()
                        squad.wcq = row["wcq"]
                        squad.date = row["date"]
                        squad.squad_home_away = row["squad_home_away"]
                        
                        if i == 0:
                            squad.squad = row["squad_home"]
                            squad.venue = "Home"
                            squad.result = row["result_home"]
                            squad.formation = row["formation_home"]
                        else:
                            squad.squad = row["squad_away"]
                            squad.venue = "Away"
                            squad.result = row["result_away"]
                            squad.formation = row["formation_away"]
                
                        for h in tr.find_all("th"):
                            if hasattr(squad, h['data-stat']):
                                setattr(squad, h['data-stat'], h.get_text().strip()) 
                            
                        for d in tr.find_all("td"):
                            if hasattr(squad, d['data-stat']):
                                setattr(squad, d['data-stat'], d.get_text().strip()) 

                        if squad.position == "GK":
                            for tr in tbody_g.find_all("tr"):
                                for d in tr.find_all("td"):
                                    if hasattr(squad, d['data-stat']):
                                        setattr(squad, d['data-stat'], d.get_text().strip())          

                        matches.append(squad)

        return pd.DataFrame([r.__dict__ for r in matches])
        
    except Exception as e:
        print("error:", e)
        print("link:", link)
        exception_type, exception_object, exception_traceback = sys.exc_info()
        filename = exception_traceback.tb_frame.f_code.co_filename
        line_number = exception_traceback.tb_lineno

        print("Exception type: ", exception_type)
        print("File name: ", filename)
        print("Line number: ", line_number)
```
